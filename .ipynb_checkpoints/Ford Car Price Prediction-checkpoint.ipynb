#### Categorical variables

```python
#check unique values for model
df['model'].unique()
```

```python
#Create another grouping or categories of the cars
def model_segment(model):
    if model.strip() == 'Fiesta' or model.strip() == 'Focus':
        return 'Compact Cars'
    elif 'Kuga' in model or 'EcoSport' in model or 'Edge' in model:
        return 'SUVs'
    elif 'Tourneo Custom' in model or 'S-MAX' in model or 'B-MAX' in model:
        return 'Minivans'
    elif 'Galaxy' in model or 'Mustang' in model or 'Grand Tourneo Connect' in model or 'Mondeo' in model:
        return 'Luxury Cars'
    elif 'Fusion' in model:
        return 'Mid-size Sedans'
    elif 'Ranger' in model:
        return 'Pick-up'
    else:
        return 'Other'
    
```

```python
df['model'] = df['model'].apply(model_segment)
```

### Data Processing

```python
# OUTLIER: Delete the rows where years > 2023
df = df[df['year'] <= 2023]
df.shape
```

```python
#check for missing value
missing = df.isnull().sum()*100/df.shape[0]
missing[missing > 0].sort_values(ascending =  False)
```

#### Label Encoding of the categorical variables

```python
# Loop over each column in the DataFrame where dtype is 'object'
for col in df.select_dtypes(include = ['object']).columns:
    print(f"{col}: {df[col].unique()}")
```

```python
from sklearn import preprocessing

#loop over each column in the Datafarame where dtype is 'object'
for col in df.select_dtypes(include = ['object']).columns:
                           
        # initialize a label encoder object
        label_encoder = preprocessing.LabelEncoder()
                           
        #Fit the encoder to the unique values in the columns
        label_encoder.fit(df[col].unique())
                           
        #Transform the column using the encoder
        df[col] = label_encoder.transform(df[col])
                           
        #Print the column name and the unique values encoded
        print(f"{col}: {df[col].unique()}")
```

### Train test split

```python
from sklearn.model_selection import train_test_split

# Perform training -  test split
X_train, X_test, y_train, y_test = train_test_split(df.drop('price', axis = 1), df['price'], test_size = 0.2, random_state = 42)
```

### Eliminating Outliers using IQR

```python
#cONCATENATE x_TRAIN AND Y_TRAIN FOR THE OUTLIER REMOVAL
train_df = pd.concat([X_train, y_train], axis = 1)

#Calculate the IQR values for each column
Q1 = train_df.quantile(0.25)
Q3 = train_df.quantile(0.75)

IQR = Q3 - Q1

#Remove outliers from X_train
train_df = train_df[~((train_df < (Q1 - 1.5*IQR)) | (train_df > (Q3 + 1.5*IQR))).any(axis = 1)]

#Separate X_train and y_train after outlier removal
X_train = train_df.drop('price', axis = 1)
y_train = train_df['price']
```

### Training the model

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import fetch_california_housing
import sklearn.metrics
from sklearn.metrics import get_scorer_names
from sklearn.metrics import accuracy_score
```

```python
# Create a DecisionTreeRegressor object
dtree = DecisionTreeRegressor()

#define the hyperparameters to tune the values
param_grid = {
    'max_depth': [2,4,6,8],
    'min_samples_split': [2,4,6,8],
    'min_samples_leaf': [1,2,3,4],
    'max_features': ['auto', 'sqrt', 'log2']
}

#Create a GridSearchCV object
scoring_metric = 'accuracy'
grid_search = GridSearchCV(dtree, param_grid, cv = 5, scoring=scoring_metric)

#Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

#Print the best hyperparameters
print(grid_search.best_params_)
```

```python
print(grid_search.best_params_)
```

```python
dtree = DecisionTreeRegressor(random_state = 0, max_depth = 2, max_features = 'auto', min_samples_leaf = 1, min_samples_split = 2)
dtree.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
from sklearn import metrics
import math

y_pred = dtree.predict(X_test)
mae = metrics.mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
rmse = math.sqrt(mse)

print('MAE is {}'.format(mae))
print('MAPE is {}'.format(mape))
print('MSE is {}'.format(mse))
print('R2 Score is {}'.format(r2))
print('RMSE is {}'.format(rmse))
```

### Feature Importance

```python
imp_df = pd.DataFrame({
    "Feature Name": X_train.columns,
    "Importance": dtree.feature_importances_
})

fi = imp_df.sort_values(by = "Importance", ascending = False)

fi2 = fi.head(10)
plt.figure(figsize = (10,8))
sns.barplot(data = fi2, x = 'Importance', y = 'Feature Name')
plt.title('Feature Importance Each Attributes (Decision Tree Regressor)', fontsize = 18)
plt.xlabel('Importance', fontsize = 16)
plt.ylabel('Feature Name', fontsize = 16)
plt.show()
```
